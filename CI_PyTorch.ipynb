```python
x = torch.arange(start = 0, end = 10, step = 1, dtype = torch.float32, requires_grad=True)
y = x**2
```

```python
y.backward(gradient=torch.ones(len(y)))
```

```python
print(x.grad)
```

```python
print(x.grad == 2*x)
```

```python
x_1 = torch.arange(start = 0, end = 10, step = 0.1, dtype = torch.float32, requires_grad=True)
y_1 = torch.sin(x_1)
```

```python
y_1.backward(gradient = torch.ones(len(y_1)))
```

```python
print(x_1.grad == torch.cos(x_1))
```

```python
plt.plot(y_1.detach().numpy(), color = "red")
plt.plot(x_1.grad.numpy(), color = "blue")
plt.show()
```

```python
x_2 = torch.arange(start = -10, end = 10, step = 0.1, dtype = torch.float32, requires_grad = True)
y_2 = relu(x_2)
```

```python
plt.plot(y_2.detach().numpy(), color = "red")
plt.plot(x_2.grad.numpy(),  color = "blue")
plt.show()
```

```python
import torchvision
from torchvision import datasets
from torchvision import transforms
from torch.utils.data import DataLoader
```

```python
train_ds = datasets.MNIST(

       root = "data",
       train = True,
       download = True,
       transform = transforms.ToTensor()
)

test_ds = datasets.MNIST(

        root = "data",
        train = False,
        download = True,
        transform = transforms.ToTensor()
)
```

```python
train_dl = DataLoader(train_ds, batch_size = 32, shuffle = True)
test_dl = DataLoader(test_ds, batch_size = 32, shuffle = False)
```

```python
images, labels = next(iter(train_dl))
out = torchvision.utils.make_grid(images)

plt.imshow(np.transpose(out, (1,2,0)))
plt.show()
```
